```python
plt.figure(figsize=(15,7))
sns.countplot(x='Season', data=Matches , order=Matches['Season'].value_counts().index.sort_values())
plt.xlabel('Season',fontsize=12)
plt.ylabel('No of matches played in each ipl season', fontsize=12)
plt.tittle('Number of matches played in each IPL season', fontsize=16)
```
